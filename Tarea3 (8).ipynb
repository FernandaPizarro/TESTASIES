```python
# reading in
import pandas as pd
import os
FigueData = pd.read_excel("/content/dataFigueroa.xlsx", index_col=0)
```

```python
FigueData.head()
```

```python
#Como se pretendía, la familia aparece como el índice de la fila (no la primera columna de datos). La familia es también los nombres de las columnas:
FigueData.columns
```

```python
varsToDrop=['Multinacional']
adjacency=FigueData.drop(varsToDrop,axis=1)

#result
adjacency
```

```python
!pip install networkx
import networkx as nx

#gráfo
EliteNet = nx.from_pandas_adjacency(adjacency)
```

```python
# plot simple
nx.draw_random(EliteNet,
                node_color='pink',
                edge_color='lightblue',
                with_labels=True,
                font_size=8)
```

```python
#Sin embargo, se refleja las autorelaciones, por lo que se quitan
EliteNet.remove_edges_from(nx.selfloop_edges(EliteNet))
```

```python
# re plot
nx.draw_random(EliteNet,
                node_color='pink',
                edge_color='lightblue',
                with_labels=True,
                font_size=8)
```

```python
FigueData['Multinacional'].head()
```

```python
#vemos que no tenemos atributos
EliteNet.nodes.data()
```

```python
dict(zip(FigueData.index,FigueData['Multinacional']))
```

```python
#Entonces, lo tomamos para un atributo del nodo
attributeToAdd=dict(zip(FigueData.index,FigueData['Multinacional']))
nx.set_node_attributes(EliteNet, attributeToAdd,'multi')
```

```python
#Elejimos color del atributo
colors={e:'red' if v==1 else 'green' for e,v in nx.get_node_attributes(EliteNet,"multi").items()}
nx.set_node_attributes(EliteNet, colors, "color")

# updated nodes
EliteNet.nodes.data()
print(EliteNet.nodes.data())
```

```python
import matplotlib.pyplot as plt
# Obtener los colores de los nodos
node_colors = [data['color'] for _, data in EliteNet.nodes(data=True)]

# Dibujar el grafo con los colores asignados
plt.figure(figsize=(10, 10))
nx.draw(EliteNet, node_color=node_colors, with_labels=True)
plt.show()
```

```python
#Plot using Pyvis
from pyvis.network import Network
import networkx as nx
import pandas as pd
```

```python
# Crear la visualización del grafo con pyvis
Gvis = Network('500px', '500px', notebook=True, cdn_resources='remote')
```

```python
# Convertir el grafo de NetworkX a pyvis
Gvis.from_nx(EliteNet)
```

```python
# Mostrar la visualización del grafo en un archivo HTML
Gvis.from_nx(EliteNet.copy()) # create a copy!
Gvis.save_graph('Gvis.html')
```

```python
Gvis.show('Gvis.html')
```

```python
#Exportarlo a Gephi
nx.write_gml(EliteNet, "Gvis.gml")
```
